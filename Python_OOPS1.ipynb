## Single Inheritance

```python

class Employee: #Parent Class
    def work(self): # parent Attribute
        print("Employee works")

class Manager(Employee):#Child Class
    def manage(self): #child Atrribute
        print("Manager manages team")

#class Your_class(Parent):
    #def your_attrib(self):
        #print("fdvveg")
        

m = Manager() #Object defining
m.work()  #Calling Parent Attribute inherited from Employee Class
m.manage()
```

## Multi-Level Inheritance

```python

class Person:  #grandparent Class
    def identity(self): #grandparent Attribute
        print("I am a person")

class Employee(Person):   #parent Class
    def job(self):    #parent Atrribute
        print("I am an employee")

class Manager(Employee):  #child Class
    def role(self):  #Child Attribute
        print("I am a manager")

m = Manager()   #object Def
m.identity()    #inherited from Grandparent Class
m.job()         #inherited from parent Class
m.role()
```

## Multiple Inheritance

```python

class Company: #parent class
    def company(self):
        print("Company name")

class Salary:  #parent Class
    def salary(self):
        print("Salary credited")

class Employee(Company, Salary): #child Class
    pass

e = Employee()
e.company()
e.salary()
```

## Encapsulation

```python
# Encapsulation - ATM Machine - we cant see money inside, only use buttons.
#private data and public methods
# Encapsulation is the concept of wrapping data and methods together and restricts direct access."

class Bank:
    def __init__(self):
        self.__balance = 0  #private Variable

    def deposit(self, amount):  #setter
        self.__balance += amount

    def get_balance(self):    #Getter
        return self.__balance

b = Bank()
b.deposit(1000)
print(b.get_balance())
```

## Abstraction

```python
# Just Rules, Not Implementation
# Abstract Base Classes (ABCs) - Define a blueprint for other classes.
# Abstraction hides implementation and enforces method structure. 
from abc import ABC, abstractmethod

class Employee(ABC):
    @abstractmethod
    def salary(self):
        pass

class Developer(Employee):
    def salary(self):
        print("Developer salary")

d = Developer()
d.salary()
```

## Property / Getter / Setter

```python
# Property Decorators - Getters and Setters
#Why needed - To control access to private attributes 
# No direct modification of private variables.
class Employee:
    def __init__(self):
        self.__salary = 0

    @property  #getter
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, amount):
        if amount > 0 :
            self.__salary = amount

e = Employee()
e.salary = 50000
print(e.salary)
```

## Polymorphism

```python
# The same Button Work works in Many forms class - Teacher - Teaches and Manager - Manages and Employee - Works
#  (Eg: Driver - Drives, Pilot - Flies etc.)
# Polymorphism allows us to use the same method to behave differently based on the object that calls it.

class Employee:
    def work(self):
        print("Employee works")

class Manager(Employee):
    def work(self):
        print("Manager manages Employees")

for e in (Employee(), Manager()):
    e.work()
```

Difference Between Abstraction and Polymorphism 
1. abstraction forces methods by defining Rules while polymorphism changes behavior
2. focus - Abstraction - what to do and polymorphism - how it behaves
3. Tool -  @abstractmethod and Poly - Method Overriding
4. Eg. - Salary() must exist and work() behaves differently

## Shallow vs Deep Copy

```python
# Shallow - shares same Toy
# deep - separate Toy
# shallow copy copies references of nested objects, deep copy creates independent copies.
import copy

a = [[1,2],[3,4]]
b = copy.copy(a)
b[0][0] = 999
print("Shallow:", a)

c = copy.deepcopy(a)
c[0][0] = 101
print("Deep:", c)
```

## Inheritance vs Composition

Inheritance builds heirarchy and Composition builds scalable systems
prefer composition over inheritance
Manager is an Employee - Car has an Engine 

```python

class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()

    def drive(self):
        self.engine.start()
        print("Car driving")

c = Car()
c.drive()
```

```python
#Dunder Methods Also Known Magic Methods
#Dunder = Double Underscore __

class Student:
    def __init__(self,name):
        self.name = name

    def __str__(self):
        return f'Student name is {self.name}'

s = Student("Anjali")
print(s)
```

```python
class Salary:
    def __init__(self,amount):
        self.amount = amount

    def __add__(self,other):
        return self.amount + other.amount

s1 = Salary(300000)
s2 = Salary(500000)

print(s1+s2)
```
